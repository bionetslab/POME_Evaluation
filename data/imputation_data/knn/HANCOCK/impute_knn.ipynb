```python
import pandas as pd
import os
from sklearn.impute import KNNImputer
import numpy as np
```

```python
# Iterate over complete simulated data directory and impute all datasets.
# Directory containing your pickle files
input_directory = 'simulated_ohe_data/'
output_directory = 'imputed_ohe_data/'

# Iterate over all files in the directory
for filename in os.listdir(input_directory):
    if filename.endswith('.tsv'):
        # Full file path
        file_path = os.path.join(input_directory, filename)
        
        # Load simulated df.
        df = pd.read_csv(file_path, sep='\t', index_col=0)
        
        # Map pandas NA to numpy NA.
        df = df.replace(pd.NA, np.nan)
        
        # Get filename without extension.
        name_without_ext = os.path.splitext(filename)[0]
        
        # Split by underscores.
        parts = name_without_ext.split('_')
        
        # Extract list of binary variable columns.
        #binary_cols = [col for col in df.columns if df[col].dropna().nunique() == 2]
        
        # Run KNN imputer.
        imputer = KNNImputer()
        imputed_res = imputer.fit_transform(df)
        imputed_df = pd.DataFrame(imputed_res, columns=df.columns, index=df.index)
        
        # Save imputed df.
        imputed_filename = os.path.join(output_directory, f'{name_without_ext}.tsv')
        imputed_df.to_csv(imputed_filename, sep='\t', index=True)
```
